<big>公式(1)<br></big>
<big>$e^{i\pi}+1=0$<br></big>
<big>$e^{i\theta}=cos\theta+isin\theta$</big><br>
<big><big>$cos\theta=\frac{e^{i\theta}+e^{-i\theta}}{2i}$</big></big><br>
<big><big>$sin\theta=\frac{e^{i\theta}-e^{-i\theta}}{2i}$</big></big>

<big>公式(2)</big><br>
<big>$x=rcos\varphi$</big><br>
<big>$y=rsin\varphi$</big><br>
<big><big>$$J_{F}(r,\varphi)=\begin{bmatrix}
{\frac{\partial x}{\partial \varphi}}&{\frac{\partial x}{\partial r}}\\
{\frac{\partial y}{\partial \varphi}}&{\frac{\partial y}{\partial r}}\\
\end{bmatrix}=\begin{bmatrix}
{-rsin\varphi}&{cos\varphi}\\
{rcos\varphi}&{sin\varphi}\\
\end{bmatrix}$$</big></big>
